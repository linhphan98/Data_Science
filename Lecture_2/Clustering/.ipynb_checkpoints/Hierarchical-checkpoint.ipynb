```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
X = df.iloc[:,[3,4]].values
```

```python
# Dendrogram to find the optimal num of clusters 
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,10))
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
```

```python
# straigthen out the horizontal bar and check which longest vertical line does not cross
# any horizontal bar. Then draw a horizontal bar crossing that vertical line and see how
# many vertical lines it crosses. That is the optimal num of clusters. 
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=5)
```

```python
y_cluster = model.fit_predict(X)
```

```python
plt.scatter(X[y_cluster==0,0],X[y_cluster==0,1],c='red',label='class 1')
plt.scatter(X[y_cluster==1,0],X[y_cluster==1,1],c='blue',label='class 2')
plt.scatter(X[y_cluster==2,0],X[y_cluster==2,1],c='black',label='class 3')
plt.scatter(X[y_cluster==3,0],X[y_cluster==3,1],c='yellow',label='class 4')
plt.scatter(X[y_cluster==4,0],X[y_cluster==4,1],c='green',label='class 5')
plt.legend()
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
```
