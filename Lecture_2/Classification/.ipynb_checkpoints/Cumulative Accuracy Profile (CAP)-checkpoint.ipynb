```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
X = df.drop('Purchased',axis=1).values
y = df['Purchased'].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
prediction = model.predict(X_test)
```

```python
accuracy_score(y_test,prediction)
```

```python
# Random CAP
```

```python
length_y_test = len(y_test)
```

```python
# counting 0 and 1 in test data 
one_count = np.sum(y_test)
zero_count = length_y_test - one_count
```

```python
# x-axis ranges from 0 to total people   
# y-axis ranges from 0 to the total positive outcomes (purchased).
plt.figure(figsize=(10,6))
plt.plot([0, length_y_test], [0, one_count], c = 'b',  
         linestyle = '--', label = 'Random Model') 
plt.legend() 
```

```python
# Random Forest Classifier Line
```

```python
lm = [y for _, y in sorted(zip(prediction, y_test), reverse = True)] 
# We extract only the y_test values in an array 
```

```python
x = np.arange(0, length_y_test + 1) 
y = np.append([0], np.cumsum(lm))
```

```python
plt.figure(figsize=(10,6))
plt.plot([0, length_y_test], [0, one_count], c = 'b',  
         linestyle = '--', label = 'Random Model') 
plt.plot(x, y, c = 'b', label = 'Random classifier', linewidth = 2) 
plt.legend() 
```

```python
# Perfect CAP 
```

```python
plt.figure(figsize=(10,6))
plt.plot([0, length_y_test], [0, one_count], c = 'b',  
         linestyle = '--', label = 'Random Model') 
plt.plot(x, y, c = 'b', label = 'Random classifier', linewidth = 2) 
plt.plot([0, one_count, length_y_test], [0, one_count, one_count], 
         c = 'grey', linewidth = 2, label = 'Perfect Model')
plt.legend() 
```

```python
# In any case, our classifier algorithm should not produce a line that lies under the random line. 
# It is considered to be a really bad model in that case. Since the plotted classifier line is close to 
#the ideal line we can say that our model is a really good fit. 
```
