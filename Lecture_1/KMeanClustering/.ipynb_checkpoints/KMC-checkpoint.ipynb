```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2,
                 centers=4, cluster_std=1.8, random_state=101)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True, figsize=(10,6))

ax1.set_title('K_means')
ax1.scatter(data[0][:,0], data[0][:,1],c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1],c=data[1], cmap='rainbow')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

## Get the Data

** Read in the College_Data file using read_csv. Figure out how to set the first column as the index.**

data = pd.read_csv('College_Data', index_col=0)

**Check the head of the data**

data.head()

** Check the info() and describe() methods on the data.**

data.info()

data.describe()

## EDA

It's time to create some data visualizations!

** Create a scatterplot of Grad.Rate versus Room.Board where the points are colored by the Private column. **

sns.scatterplot(data['Room.Board'], data['Grad.Rate'], hue=data['Private'])

**Create a scatterplot of F.Undergrad versus Outstate where the points are colored by the Private column.**

sns.scatterplot(data['Outstate'], data['F.Undergrad'], hue=data['Private'])

** Create a stacked histogram showing Out of State Tuition based on the Private column. Try doing this using [sns.FacetGrid](https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.FacetGrid.html). If that is too tricky, see if you can do it just by using two instances of pandas.plot(kind='hist'). **

sns.set_style('darkgrid')
g = sns.FacetGrid(data, hue='Private', aspect = 1.5, palette='coolwarm')
g = g.map(plt.hist, 'Outstate', bins=20)

**Create a similar histogram for the Grad.Rate column.**

sns.set_style('darkgrid')
g = sns.FacetGrid(data, hue='Private', aspect = 2, palette='coolwarm')
g = g.map(plt.hist, 'Grad.Rate', bins = 20)

** Notice how there seems to be a private school with a graduation rate of higher than 100%.What is the name of that school?**

data[data['Grad.Rate'] > 100]

** Set that school's graduation rate to 100 so it makes sense. You may get a warning not an error) when doing this operation, so use dataframe operations or just re-do the histogram visualization to make sure it actually went through.**



data[data['Grad.Rate'] > 100] = 100

sns.set_style('darkgrid')
g = sns.FacetGrid(data, hue='Private', aspect = 2, palette='coolwarm')
g = g.map(plt.hist, 'Grad.Rate', bins = 20)

## K Means Cluster Creation

Now it is time to create the Cluster labels!

** Import KMeans from SciKit Learn.**

from sklearn.cluster import KMeans

# ** Create an instance of a K Means model with 2 clusters.**

kmean = KMeans(n_clusters=2)

# **Fit the model to all the data except for the Private label.**

kmean.fit(data.drop('Private', axis=1))

# ** What are the cluster center vectors?**

kmean.cluster_centers_

## Evaluation

# There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.

# ** Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

cluster = pd.get_dummies(data['Private'], drop_first=True)
cluster.drop('No', axis = 1, inplace=True)

data = data.join(cluster)

data.head()

# ** Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(data['Yes'], kmean.labels_))
print(classification_report(data['Yes'], kmean.labels_))
```
