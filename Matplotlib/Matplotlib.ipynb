```python
import matplotlib.pyplot as plt

%matplotlib inline 

#see plot created inside notebook

import numpy as np
x = np.linspace(0,5,11)
y = x**2

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')

plt.subplot(1,2,1) # row, col, first plot
plt.plot(x,y,'r')

plt.subplot(1,2,2) # row, col, second plot
plt.plot(y,x,'b')

# object oriented method
fig = plt.figure() 

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title('Smaller Plot')
axes1.set_title('Larger Plot')

fig, axes = plt.subplots(nrows = 1, ncols = 2)
# plt.tight_layout()

# for cur in axes:
#     cur.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')



fig = plt.figure(figsize=(8,4)) # width, height

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8,4))
plt.tight_layout()

axes[0].plot(x,y)
axes[1].plot(y,x)

fig.savefig('my_pic.png')



fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x**2, label='X Squared')
ax.plot(x**3, label='X Cubed')

ax.legend(loc=(0.02,0.85))



fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='r', lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
